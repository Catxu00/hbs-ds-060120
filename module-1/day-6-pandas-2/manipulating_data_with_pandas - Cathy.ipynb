### 2.  Changing data

#### DataFrame.applymap() and Series.map()

The ```.applymap()``` method takes a function as input that it will then apply to every entry in the dataframe.

````python
def successor(x):
    return x + 1
````

````python
uci.applymap(successor).head()
````

The `.map()` method takes a function as input that it will then apply to every entry in the Series.

````python
uci['age'].map(successor).tail(10)
````

````python
# instead of passing in a function, pass in a dict:
uci['sex'].map({0:'male',1:'female'}).head()
````

#### Anonymous Functions (Lambda Abstraction)

Simple functions can be defined right in the function call. This is called 'lambda abstraction'; the function thus defined has no name and hence is "anonymous".

````python
uci['oldpeak'].map(lambda x: round(x))[:4]
````

````python
# above is same as:
def new_round_func(x):
    return round(x)
````

````python
uci['oldpeak'].map(new_round_func)[:4]
````

````python
uci['oldpeak'].map(lambda x: f'my new peak is {x}!')[:4]
````

````python
uci['new peak'] = uci['oldpeak'].map(lambda x: f'my new peak is {x}!')[:4]
````

````python
uci.head()
# row 4 is NaN because the code above is only applied to first 4 rows. 
````

````python
# Second, there are four units: year, month, week, day. 
def convert_to_days_old(val):
    number, unit = val.split(' ')
    if 'year' in val:
        return 365 * number
    if 'month' in unit:
        return 30 * number
    if 'week' in unit:
        return 7 * number
    if 'day' in unit:
        return number
    
    return 'unknown'
````

````python
animal_outcomes.age_upon_outcome.map(convert_to_days_old)
````

````python
# ^This happened becasue .split() is a string function, so multiplication returns more strings. 
````

````python
def convert_to_days_old(val):
    number, unit = val.split(' ')
    number = int(number)
    if 'year' in val:
        return 365 * number
    if 'month' in unit:
        return 30 * number
    if 'week' in unit:
        return 7 * number
    if 'day' in unit:
        return number
    
    return 'unknown'
````

````python
animal_outcomes['days_upon_outcome'] = animal_outcomes.age_upon_outcome.map(convert_to_days_old)
````

### Methods for Combining DataFrames: `.join()`, `.merge()`, `.concat()`, `.melt()`

### `.join()`

````python
toy1 = pd.DataFrame([[63, 142], [33, 47]], columns = ['age', 'HP'])
toy2 = pd.DataFrame([[63, 100], [33, 200]], columns = ['age', 'HP'])
````

````python
toy1.join(toy2.set_index('age'),
          on = 'age',
          lsuffix = '_A',
          rsuffix = '_B').head()
````

### `.merge()`

````python
ds_chars = pd.read_csv('data/ds_chars.csv', index_col = 0)
````

````python
states = pd.read_csv('data/states.csv', index_col = 0)
````

````python
ds_chars.merge(states,
               left_on='home_state',
               right_on = 'state',
               how = 'inner')
````
